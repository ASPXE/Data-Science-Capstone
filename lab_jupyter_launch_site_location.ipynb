```python
# TASK 3: Calculate the distances between a launch site to its proximities

# Function to calculate distance between two geographical points
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Earth radius in kilometers
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Example: Let's assume we've manually identified a point on the coastline near one of the launch sites
coastline_lat = 28.56367  # Latitude of coastline
coastline_lon = -80.57163  # Longitude of coastline

# Example: Let's calculate the distance from one of the launch sites to the coastline
launch_site_lat = 28.573255  # Latitude of the launch site
launch_site_lon = -80.646895  # Longitude of the launch site

distance_to_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance to coastline: {distance_to_coastline:.2f} km")

# Now let's add the marker and distance line to the map
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=folium.DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12; color:#d35400;"><b>{distance_to_coastline:.2f} KM</b></div>'
    )
)

# Add the marker to the map
site_map.add_child(distance_marker)

# Draw a line between the launch site and the coastline
line = folium.PolyLine(
    locations=[(launch_site_lat, launch_site_lon), (coastline_lat, coastline_lon)],
    color='blue',
    weight=2
)

# Add the line to the map
site_map.add_child(line)

# Display the updated map with the distance line
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
from math import sin, cos, sqrt, atan2, radians
import folium
from folium.features import DivIcon

# Define the function to calculate the distance using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # radius of Earth in kilometers
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c  # distance in kilometers
    return distance

# Example launch site (replace with actual values)
launch_site_lat = 28.573255
launch_site_lon = -80.646895

# Example proximity (e.g., city, railway, or coastline) (replace with actual values)
proximity_lat = 28.56125  # e.g., a city's latitude
proximity_lon = -80.57736  # e.g., a city's longitude

# Calculate the distance between launch site and proximity
distance = calculate_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)

# Create a folium map centered around the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Create a marker for the launch site
launch_marker = folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color='blue', icon='cloud')
)
launch_marker.add_to(site_map)

# Create a marker for the proximity (e.g., city, railway, or coastline)
proximity_marker = folium.Marker(
    location=[proximity_lat, proximity_lon],
    popup=f"Proximity (Distance: {distance:.2f} km)",
    icon=folium.Icon(color='red', icon='info-sign')
)
proximity_marker.add_to(site_map)

# Create a PolyLine to draw a line between the launch site and the proximity
line = folium.PolyLine(
    locations=[(launch_site_lat, launch_site_lon), (proximity_lat, proximity_lon)],
    color='green',
    weight=2
)
line.add_to(site_map)

# Add a DivIcon to display the distance near the proximity marker
distance_marker = folium.Marker(
    location=[proximity_lat, proximity_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12; color:#d35400;"><b>{distance:.2f} km</b></div>'
    )
)
distance_marker.add_to(site_map)

# Display the map
site_map
```
